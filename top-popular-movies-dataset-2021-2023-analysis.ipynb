```python
import pandas as pd
```

```python
#Reading The CSV File

data = pd.read_csv("top_popular_movies.csv")
```

```python
#Checking Our Dataset

data.head(n=20)
```

```python
#Checking For Any Null Values

data.isnull().sum()
```

```python
#Droping The Rows

data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
data.head(20)
```

```python
data.shape
```

```python
#Checking Most Movies Per Rating

data['vote_average'].value_counts().sort_values()
```

```python
#Checking Most Movies Per Rating

import matplotlib.pyplot as plt

movies_per_rating = data['vote_average'].value_counts()

plt.bar(movies_per_rating.index, movies_per_rating.values)
plt.title("Movies Based Per Rating")
plt.xlabel("Ratings")
plt.ylabel("Movies")
plt.show()
```

```python
#Checking Movies With The Rating 7 and Above 

from tabulate import tabulate

high_rated_movies = data[(data['vote_average'] >= 7.0)]

# create a list of lists containing the title and rating for each movie
movie_data = []
for index, row in high_rated_movies.iterrows():
    movie_data.append([row['title'], row['vote_average']])

# format the movie data as a table and print it out
print(tabulate(movie_data, headers=['Title', 'Vote_Average']))
```

```python
from wordcloud import WordCloud


# extract the movie titles as a single string
movie_titles = ' '.join(data['title'].tolist())

# create a wordcloud from the movie titles
wordcloud = WordCloud(background_color='black').generate(movie_titles)

# plot the wordcloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
